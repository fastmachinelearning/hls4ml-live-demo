```python
# Import the driver Overlay class
from axi_stream_driver import NeuralNetworkOverlay
```

```python
import numpy as np
X_test = np.load('X_test_NRN.npy')
y_test = np.load('y_test_NRN.npy')
X_test = X_test[:50]*256
y_test = y_test[:50]
```

```python
nn = NeuralNetworkOverlay('NewBranch.bit', X_test[0].shape, y_test[0].shape)
```

```python
from pynq.lib.video import *
#from pynq.lib.video import hierarchies
```

```python
# monitor configuration: 640*480 @ 60Hz
Mode = VideoMode(640,480,24)
#hdmi_out = base.video.hdmi_out
hdmi_out = nn.video.hdmi_out
hdmi_out.configure(Mode,PIXEL_BGR)
hdmi_out.start()
```

```python
# initialize camera from OpenCV
import cv2 as cv

videoIn = cv.VideoCapture(-1)
while(videoIn.isOpened() == False):
    videoIn = cv.VideoCapture(-1)
    videoIn.set(cv.CAP_PROP_FRAME_WIDTH, 640);
    videoIn.set(cv.CAP_PROP_FRAME_HEIGHT, 480);

print("Capture device is open: " + str(videoIn.isOpened()))
```

```python
import time
start = time.time()
for NumOfFrames in range (50): 
    ret, frame = videoIn.read()
    if (ret):
        outframe = hdmi_out.newframe()
        outframe[0:480,0:640,:] = frame[0:480,0:640,:]        
        hdmi_out.writeframe(outframe)
    else:
        raise RuntimeError("Failed to read from camera.")
end = time.time()
print("Frames per second:     " + str(50 / (end - start)))
```

```python
# Capture webcam video and display to HDMI Output
import numpy as np
x,y,w,h = 300,0,150,40
import time
start = time.time()
for NumOfFrames in range (500): 
    ret, frame = videoIn.read()
    if (ret):
        outframe = hdmi_out.newframe()
        cropped_frame = frame[0:480, 0:480]
        resized = cv.resize(cropped_frame, (32, 32))
        y_hw, latency, throughput = nn.predict(resized, debug=False, profile=True)
        percentage = np.array(y_hw)
        percentage_max = np.argmax(percentage)
        if percentage_max == 0:
            text = "Class A"
        elif percentage_max == 1: 
            text = "Class B"
        elif percentage_max == 2:
            text = "Class C"
        elif percentage_max == 3: 
            text = "Class D"
        elif percentage_max == 4: 
            text = "Class E"
        elif percentage_max == 5: 
            text = "Class F"
        elif percentage_max == 6: 
            text = "Class G"
        elif percentage_max == 7: 
            text = "Class H"
        elif percentage_max == 8: 
            text = "Class I"
        elif percentage_max == 9: 
            text = "Class J"
        else: 
            text = "Class K"
        print(text)
        outframe[0:480,0:640,:] = frame[0:480,0:640,:]        
        cv.rectangle(outframe, (x,y), (x + w, y + h), (0,0,0), -1)
        cv.putText(img=outframe, text=text,org=(x+int(w/10),y+int(4*h/5)), fontFace=cv.FONT_HERSHEY_DUPLEX, fontScale=1, color=(255,0,0), thickness=1)
        #outframe[0:32,0:32,:] = crop_frame[0:32,0:32,:]
        #outframe = cv.resize(crop_frame,None,fx=(1/15),fy=(1/15),interpolation=cv.INTER_AREA)
        hdmi_out.writeframe(outframe)
    else:
        raise RuntimeError("Failed to read from camera.")
end = time.time()
print("Frames per second:     " + str(50 / (end - start)))
```
